To prepare the dataset for training and limit its size, we'll first load the image paths and corresponding text labels from the `VinText` directory. We'll use the `datasets` library to create a `Dataset` object, which can then be easily manipulated (e.g., limited in size).

This code assumes that image files are in `VinText/vietnamese/images` (e.g., `img_123.jpg` or `123.png`) and label files are in `VinText/vietnamese/labels` (e.g., `gt_123.txt`). If this structure is different, you might need to adjust the `image_pattern`, `label_pattern`, and the logic for matching image and label IDs.

```python
import os
import glob
import re
from datasets import Dataset

# Define base paths and file patterns
base_path = "VinText/vietnamese"
image_pattern = os.path.join(base_path, "images", "*.{png,jpg}") # Covers both .png and .jpg
label_pattern = os.path.join(base_path, "labels", "gt_*.txt")

# Collect file paths
image_files = glob.glob(image_pattern)
label_files = glob.glob(label_pattern)

print(f"Found {len(image_files)} image files and {len(label_files)} label files.")

# Check if files were found
if not image_files or not label_files:
    print("\nCould not find image or label files in the expected structure.")
    print(f"Please verify if the `VinText/vietnamese/images` and `VinText/vietnamese/labels` directories exist and contain files.")
    print("You can run `!ls -R VinText` to inspect the full directory structure.")

    # Creating a dummy dataset for demonstration of limiting if real data isn't found
    print("\nCreating a dummy dataset for demonstration of dataset limiting:")
    data_items = []
    for i in range(250):
        data_items.append({"image_path": f"dummy_img_{i}.png", "text": f"This is dummy text for image {i}."})
    dataset = Dataset.from_list(data_items)

else:
    # Match image and label files based on their IDs
    label_id_to_path = {re.search(r'gt_(\d+)\.txt$', os.path.basename(f)).group(1): f for f in label_files if re.search(r'gt_(\d+)\.txt$', os.path.basename(f))}
    image_id_to_path = {}
    for img_f in image_files:
        img_basename = os.path.basename(img_f)
        match = re.match(r"(?:img_)?(\d+)\.(?:png|jpg)", img_basename)
        if match:
            image_id_to_path[match.group(1)] = img_f

    data_items = []
    for label_id, label_path in label_id_to_path.items():
        image_path = image_id_to_path.get(label_id)
        if image_path:
            with open(label_path, "r", encoding="utf-8") as f:
                text_content = f.read().strip()
            data_items.append({"image_path": image_path, "text": text_content})

    if not data_items:
        print("\nNo matching image-label pairs found. Please check file naming conventions.")
        # Fallback to dummy data
        print("Creating a dummy dataset for demonstration of dataset limiting:")
        data_items = []
        for i in range(250):
            data_items.append({"image_path": f"dummy_img_{i}.png", "text": f"This is dummy text for image {i}."})

    dataset = Dataset.from_list(data_items)

print(f"\nOriginal dataset size: {len(dataset)}")

# Set a limit for the dataset
# You can change this value to your desired limit
limit = 100

if len(dataset) > limit:
    dataset = dataset.select(range(limit))
    print(f"Dataset limited to: {len(dataset)} examples.")
else:
    print(f"Dataset size ({len(dataset)}) is already less than or equal to the limit ({limit}), no changes applied.")

print("\nFirst item in the dataset after potential limiting:")
if len(dataset) > 0:
    print(dataset[0])
else:
    print("Dataset is empty.")
```
